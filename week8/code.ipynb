```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fu = pd.read_csv('fu.csv', index_col=0)
fu
```

```python
# Less active traders
cluster_0 = fu[fu['cluster'] == 0]
c0_id = cluster_0.index.tolist()
print("num: ",len(c0_id))
c0_id
```

```python
# Stable traders

cluster_1 = fu[fu['cluster'] == 1]
c1_id = cluster_1.index.tolist()
print("num: ",len(c1_id))
c1_id
```

```python
# Dormant traders

cluster_2 = fu[fu['cluster'] == 2]
c2_id = cluster_2.index.tolist()
print("num: ",len(c2_id))
c2_id
```

```python
# High value traders

cluster_3 = fu[fu['cluster'] == 3]
c3_id = cluster_3.index.tolist()
print("num: ",len(c3_id))
c3_id
```

```python
# High value traders
df_High_value_traders = data[data['Identifier'].isin(c3_id)]
df_High_value_traders
```

```python
df_High_value_traders_Weekly = df_High_value_traders.groupby('Week Number').mean()
df_High_value_traders_Weekly = df_High_value_traders_Weekly.drop("Identifier", axis=1)
df_High_value_traders_Weekly = df_High_value_traders_Weekly.reset_index()
df_High_value_traders_Weekly
```

---
### Stable traders
---

```python
# Stable traders
df_Stable_traders = data[data['Identifier'].isin(c1_id)]
df_Stable_traders
```

```python
df_Stable_traders_Weekly = df_Stable_traders.groupby('Week Number').mean()
df_Stable_traders_Weekly = df_Stable_traders_Weekly.drop("Identifier", axis=1)
df_Stable_traders_Weekly = df_Stable_traders_Weekly.reset_index()
df_Stable_traders_Weekly
```

---
### Less active traders
---

```python
# Less active traders
df_Less_active_traders = data[data['Identifier'].isin(c0_id)] 
df_Less_active_traders
```

```python
df_Less_active_traders_Weekly = df_Less_active_traders.groupby('Week Number').mean()
df_Less_active_traders_Weekly = df_Less_active_traders_Weekly.drop("Identifier", axis=1)
df_Less_active_traders_Weekly = df_Less_active_traders_Weekly.reset_index()
df_Less_active_traders_Weekly
```

---
### Dormant traders
---

```python
# Dormant traders
df_Dormant_traders = data[data['Identifier'].isin(c2_id)]
df_Dormant_traders
```

```python
df_Dormant_traders_Weekly = df_Dormant_traders.groupby('Week Number').mean()
df_Dormant_traders_Weekly = df_Dormant_traders_Weekly.drop("Identifier", axis=1)
df_Dormant_traders_Weekly = df_Dormant_traders_Weekly.reset_index()
df_Dormant_traders_Weekly
```

```python
# 'trading_volumes_7days'
buy = pd.merge(df_High_value_traders_Weekly[['Week Number', 'buy_number_7days']], df_Stable_traders_Weekly[['Week Number', 'buy_number_7days']],
                     on='Week Number', how='outer')

buy = pd.merge(buy, df_Less_active_traders_Weekly[['Week Number', 'buy_number_7days']],
                     on='Week Number', how='outer')


buy = pd.merge(buy, df_Dormant_traders_Weekly[['Week Number', 'buy_number_7days']],
                     on='Week Number', how='outer')

buy.columns = ['Week Number', 'High_value_traders', 'Stable_traders', 'Less_active_traders', 'Dormant_traders']
buy.fillna(0, inplace=True)
buy.sort_values(by='Week Number', ascending=True, inplace=True)

buy
```

```python
sell = pd.merge(df_High_value_traders_Weekly[['Week Number', 'sell_values_7days']], df_Stable_traders_Weekly[['Week Number', 'sell_values_7days']],
                     on='Week Number', how='outer')

sell = pd.merge(sell, df_Less_active_traders_Weekly[['Week Number', 'sell_values_7days']],
                     on='Week Number', how='outer')


sell = pd.merge(sell, df_Dormant_traders_Weekly[['Week Number', 'sell_values_7days']],
                     on='Week Number', how='outer')

sell.columns = ['Week Number', 'High_value_traders', 'Stable_traders', 'Less_active_traders', 'Dormant_traders']
sell.fillna(0, inplace=True)
sell.sort_values(by='Week Number', ascending=True, inplace=True)
sell
```

```python
plt.figure(figsize=(10, 6))

for column in buy.columns[1:]:
    plt.plot(buy['Week Number'], buy[column] + sell[column], label=column)

plt.legend()

plt.xlim(15,87)
plt.title('trading_volumes_7days')
plt.xlabel('Week Number')
plt.ylabel('Value')

plt.show()
```
